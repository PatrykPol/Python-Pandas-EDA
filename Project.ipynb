```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r"C:\Users\patry\OneDrive\Pulpit\sample_-_superstore.xls"

df = pd.read_excel(file_path, index_col=False, sheet_name=0)

df
```

```python
# Handling Missing Values: Filling with "Unknown"
df['Category'].fillna('Unknown', inplace=True)
```

```python
# Correcting Data Types: Converting 'Order Date' back to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Handling Outliers: Removing the specific outlier
df = df[df['Sales'] < 50000]  # Assuming normal sales are below 50,000
```

```python
# Fixing Inconsistent Category Names
df['Category'] = df['Category'].replace({'Furnishing': 'Furniture', 'Office-Supplies': 'Office Supplies'})
```

```python
# Removing Duplicate Rows
df_clean = df.drop_duplicates()
```

```python
# Removing special characters from 'Customer Name'
df['Customer Name'] = df['Customer Name'].str.replace('[^a-zA-Z ]', '', regex=True)
```

```python
# Imputing missing values in 'Sales' with the median sales value
sales_median = df['Sales'].median()
df['Sales'].fillna(sales_median, inplace=True)
```

```python
# Displaying the cleaned data to verify corrections
print(df_clean[['Order Date', 'Category', 'Sales']].head(15))
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Aggregate sales data by month
df['Order Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Order Month')['Sales'].sum().reset_index()
monthly_sales['Order Month'] = monthly_sales['Order Month'].dt.to_timestamp()

# Plotting sales trends over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Month', y='Sales', data=monthly_sales, marker='o', color='blue', linewidth=2.5)
plt.title('Sales Trends Over Time', fontsize=16)
plt.xlabel('Order Month', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Aggregate profit data by Category and Sub-Category
category_profit = df.groupby(['Category', 'Sub-Category'])['Profit'].sum().reset_index()

# Plotting profitability by category and sub-category
plt.figure(figsize=(14, 8))
sns.barplot(x='Profit', y='Sub-Category', hue='Category', data=category_profit, palette='viridis')
plt.title('Profitability by Category and Sub-Category', fontsize=16)
plt.xlabel('Total Profit', fontsize=14)
plt.ylabel('Sub-Category', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Aggregate sales data by City
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(20)

# Plotting geographical distribution of sales
plt.figure(figsize=(10, 8))
city_sales.plot(kind='bar', color='skyblue')
plt.title('Top 20 Cities by Sales', fontsize=16)
plt.xlabel('City', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Counting the frequency of each shipping mode
ship_mode_counts = df['Ship Mode'].value_counts()

# Plotting shipping mode preference
plt.figure(figsize=(10, 6))
ship_mode_counts.plot(kind='bar', color='coral')
plt.title('Shipping Mode Preference', fontsize=16)
plt.xlabel('Ship Mode', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Aggregating average discount for each quantity level
avg_discount_by_quantity = df.groupby('Quantity')['Discount'].mean().reset_index()

# Plotting only the average discount by quantity for simplicity
plt.figure(figsize=(12, 7))

# Line plot for average discount by quantity
sns.lineplot(x='Quantity', y='Discount', data=avg_discount_by_quantity, marker='o', color='red', linewidth=2.5)
plt.title('Average Discount per Order Quantity', fontsize=16)
plt.xlabel('Quantity', fontsize=14)
plt.ylabel('Discount', fontsize=14)
plt.grid(True)  # Adding grid for better readability
plt.tight_layout()
plt.show()
```

```python
# Aggregating sales data by Category and Region
category_region_sales = df.groupby(['Category', 'Region'])['Sales'].sum().unstack()

# Plotting the heatmap with a yellow-orange-red color scheme
plt.figure(figsize=(10, 6))
sns.heatmap(category_region_sales, annot=True, fmt=".0f", cmap='YlOrRd', linewidths=.5)
plt.title('Sales by Product Category and Region', fontsize=16)
plt.xlabel('Region', fontsize=14)
plt.ylabel('Category', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Selecting a subset of columns for the pair plot
subset_data = df[['Sales', 'Profit', 'Quantity', 'Discount']]

# Creating the pair plot & setting the visual style
sns.set(style="ticks", palette="pastel")
pair_plot = sns.pairplot(subset_data)

# Title with a slight adjustment in y for spacing
pair_plot.fig.suptitle('Pair Plot of Sales, Profit, Quantity, and Discount', y=1.02)

plt.show()
```
